```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading Data
mp = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

```python
#Describing Data
mp.head()
```

```python
mp.nunique()
```

```python
mp.describe()
```

```python
mp.info()
```

```python
mp = mp.dropna()
```

```python
mp.info()
```

```python
# Data Visualization
sns.pairplot(mp, x_vars=['displacement', 'horsepower', 'weight','acceleration', 'mpg'], y_vars=['mpg'])
```

```python
sns.regplot(x='displacement', y='mpg', data=mp)
```

```python
# Data Preprocessing
mp.columns
```

```python
# Define Target Variable (y) and Feature Variables (X)
y = mp['mpg']
```

```python
X = mp[['displacement', 'horsepower', 'weight',
       'acceleration']]
```

```python
X.shape, y.shape
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,train_size = 0.8 , random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Modeling
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# Model Evaluation
model.intercept_
model.coef_
model.score(X_test, y_test)
```

```python
# Prediction
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Explaination
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
